```python
img = np.array([
    [1, 2, 3, 50, 52, 52, 1, 1, 1, 1, 1, 1, 1, 1, 4, 4, 4, 4, 4],
    [51, 52, 53, 22, 23, 24, 1, 1, 1, 1, 1, 1, 1, 1, 4, 4, 4, 4, 4],
    [52, 53, 54, 108, 109, 110, 1, 1, 1, 1, 1, 1, 1, 1, 4, 4, 4, 4, 4],
    [0, 1, 1, 49, 50, 51, 1, 1, 1, 1, 1, 1, 1, 1, 4, 4, 4, 4, 4],
    [50, 51, 52, 21, 22, 23, 1, 1, 1, 1, 1, 1, 1, 1, 4, 4, 4, 4, 4],
    [51, 52, 53, 107, 106, 105, 1, 1, 1, 1, 1, 1, 1, 1, 4, 4, 4, 4, 4],
    [1] * 14 + [4] * 5,
    [1] * 14 + [4] * 5,
    [1] * 14 + [4] * 5,
    [1] * 14 + [4] * 5,
    [1] * 14 + [4] * 5,
    [1] * 14 + [4] * 5,
    [1] * 14 + [4] * 5,
    [200] * (14 + 5),
    [200] * (14 + 5),
    [200] * (14 + 5),
    [200] * (14 + 5)
], dtype=np.long)
img
```

```python
hist = np.bincount(img.flatten(), minlength=256)
hist
```

Capacities tell how much space is left in each bin

```python
clip_limit = 2
hist = hist.astype(np.int64, copy=True)
clipped = np.minimum(hist, clip_limit)
clipped
```

```python
capacities = clip_limit - clipped

while clipped[clipped < clip_limit].size > 0:
    eligible = np.where(capacities > 0)[0]
    excess = int(hist.sum() - clipped.sum())
    q, r = divmod(excess, clipped[clipped < clip_limit].size)

    if q > 0:
        clipped[eligible] += np.minimum(capacities[eligible], q)
        capacities = clip_limit - clipped
    elif r > 0:
        np.add.at(clipped, eligible[:r], 1)
    else:
        break

hist = clipped
```

```python
high = 200
low = 50
cdf = np.cumsum(hist)
cdf_min = cdf[cdf > 0][0]
cdf_norm = (cdf - cdf_min) / (cdf.max() - cdf_min)
lookup_table = (cdf_norm * (high - low) + low).astype(np.uint8)
img_eq = lookup_table[img]
img_eq
```
