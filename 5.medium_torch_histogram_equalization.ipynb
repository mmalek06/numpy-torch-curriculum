```python
import torch
```

```python
img = torch.tensor([
    [50, 52, 52],
    [51, 52, 53],
    [52, 53, 54]
], dtype=torch.uint8)
img
```

```python
hist = img.flatten().bincount(minlength=256)
hist
```

```python
cdf = hist.cumsum(dim=0)
cdf
```

```python
cdf_min = cdf[cdf > 0].min()
cdf_min
```

```python
cdf_normalized = ((cdf - cdf_min) * 255 / (cdf.max() - cdf_min)).to(torch.uint8)
cdf_normalized
```

```python
flat = img.flatten()
mask = torch.zeros(cdf_normalized.shape[0], dtype=torch.uint8)
mask[flat.unique().long()] = 1

mask.nonzero().squeeze()
```
