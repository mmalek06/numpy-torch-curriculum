```python
import torch
```

```python
img = torch.tensor([
    [50, 52, 52],
    [51, 52, 53],
    [52, 53, 54]
], dtype=torch.uint8)
img
```

```python
hist = img.flatten().bincount(minlength=256)
hist
```

```python
cdf = hist.cumsum(dim=0)
cdf
```

```python
high = 200
low = 50
cdf_norm = cdf / cdf.max()
lookup_table = (cdf_norm * (high - low) + low).long()
lookup_table
```

```python
img_eq = lookup_table[img.long()]
img_eq
```
